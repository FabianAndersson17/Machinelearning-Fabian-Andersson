# Polynomial regression
- underfit
- overfit

```python
import numpy as np
import matplotlib.pyplot as plt

samples = 100
X = np.random.uniform(-3,3, (samples, 1))
f = lambda x: x**2 + .5*x + 3
y = f(X) + np.random.randn(samples, 1)

plt.plot(X, y, ".")

x = np.linspace(-3,3)
plt.plot(x, f(x))
plt.title("Simulated data")
```

## Polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_features = PolynomialFeatures(degree = 2, include_bias=False)
poly_X = polynomial_features.fit_transform(X)
poly_X[:3]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(poly_X, y)
model.coef_
```

```python
x = np.linspace(-3,3, samples).reshape(samples, 1)

y_pred = model.predict(poly_X)
y_pred.shape, x.shape

plt.plot(X, y_pred, ".")
```
