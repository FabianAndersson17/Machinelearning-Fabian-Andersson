```python
mpg_dataset = mpg_dataset.drop(["model_year", "origin", "name", "acceleration"], axis="columns")
mpg_dataset.describe()
```

```python
fig, ax = plt.subplots(2, 3, figsize = (15,10), dpi=(100))
for i, feature in enumerate(["horsepower", "acceleration", "cylinders"]): ## Add enumerate. Automate more. Plot rest of data
    sns.scatterplot(data=mpg_dataset, x="weight", y = feature, ax=ax[0, i])

for i, feature in enumerate(["weight", "horsepower", "displacement"]):
    sns.scatterplot(data=mpg_dataset, x=feature, y="mpg", ax=ax[1, i])
```
